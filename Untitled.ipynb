```python
# –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É —Å –∑–∞–ø—Ä–æ—Å–∞–º–∏ –∏ –æ—Ç–≤–µ—Ç–∞–º–∏
input_path = "data/valid_dataset.json"

# –ó–∞–≥—Ä—É–∂–∞–µ–º –∑–∞–ø—Ä–æ—Å—ã
with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)


def generate_predictions(qa_pipeline, samples):
    scores = []

    for sample in tqdm(samples, desc="–ì–µ–Ω–µ—Ä–∞—Ü–∏—è –æ—Ç–≤–µ—Ç–æ–≤", ncols=100):
        sample["pred_response"] = qa_pipeline.invoke(sample["query"])["result"]
        score = meteor_score([sample["response"].split()], sample["pred_response"].split())
        sample["meteor"] = score
        scores.append(score)

    avg_score = np.mean(scores)
    return avg_score

def show_predictions(samples, n=15):
    """
    –û—Ç–æ–±—Ä–∞–∂–∞–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ —Å –ø–æ–¥—Å–≤–µ—Ç–∫–æ–π METEOR –∏ —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ–º.
    """
    def highlight_score(score):
        if score >= 0.75:
            return "üü¢"
        elif score >= 0.5:
            return "üü°"
        else:
            return "üî¥"

    rows = []
    for s in samples[:n]:
        rows.append({
            "üí¨ –í–æ–ø—Ä–æ—Å": s["query"],
            "üéØ –≠—Ç–∞–ª–æ–Ω–Ω—ã–π –æ—Ç–≤–µ—Ç": s["response"],
            "ü§ñ –û—Ç–≤–µ—Ç LLM": s["pred_response"],
            "üìè METEOR": f'{s["meteor"]:.3f} {highlight_score(s["meteor"])}'
        })

    df = pd.DataFrame(rows)
    display(HTML(df.to_html(escape=False, index=False)))

    return df
```
